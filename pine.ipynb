```python
// This source code is subject to the terms of the Mozilla Public License 2.0 at https://mozilla.org/MPL/2.0/


//@version=5


multi=1



fast = 13
slow = 34
atr=ta.atr(6)
fastMA = ta.ema(close, fast)
slowMA = ta.ema(close, slow)
macd = fastMA - slowMA
signal = ta.ema(macd, 9)
histLine = macd-signal
upOrDown = 0
i=1
if histLine[0] >=0
    upOrDown := 1
    while histLine[i] > 0
        i += 1
        
else
    upOrDown := -1
    while histLine[i] < 0
        i += 1
        
changebar=0

temp = 0.0
closetemp = 0.0 
if upOrDown == 1
    for j=i to 1
        if histLine[j] > histLine[j-1] and histLine[j+1]<histLine[j]
            if temp < histLine[j]
                temp := histLine[j] 
                closetemp :=close[j]
                changebar := 1
            else
                if close[j] > closetemp
                    closetemp :=close[j]
                    temp := histLine[j] 
                    changebar +=1
                else 
                    temp := histLine[j] 
                    closetemp :=close[j]
                    changebar := 1
    
else
    for j=i to 1
        if histLine[j] < histLine[j-1] and histLine[j+1]>histLine[j]
            if temp > histLine[j]
                temp := histLine[j] 
                changebar := -1
            else
                if close[j] < closetemp
                    closetemp :=close[j]
                    temp := histLine[j] 
                    changebar -=1
                else 
                    temp := histLine[j] 
                    closetemp :=close[j]
                    changebar := -1
strategy("Mace strategy")
varip ipAtr =atr
varip ipLow=low
varip ipHigh=high
varip ipClose=close
if strategy.position_size == 0
    strategy.entry("S", strategy.short,10/open, when=(changebar == 3 and changebar[1] != 3) or (changebar == 4 and changebar[1] != 4) )
    strategy.entry("L", strategy.long, 10/open, when= (changebar == -3 and changebar[1] != -3) or (changebar == -4 and changebar[1] != -4))
    ipAtr := atr
    ipLow :=low
    ipHigh :=high
    ipClose :=close
    if (changebar == 3 and changebar[1] != 3) or (changebar == 4 and changebar[1] != 4) 
        strategy.exit("S", stop=ipHigh+ipAtr, limit=ipClose-(ipHigh-ipClose+ipAtr)*multi)
        
    if (changebar == -3 and changebar[1] != -3) or (changebar == -4 and changebar[1] != -4)
        strategy.exit("L", stop=ipLow-ipAtr, limit=ipClose+(ipClose-ipLow+ipAtr)*multi)
if strategy.position_size > 0
    strategy.exit("L", stop=ipLow-ipAtr, limit=ipClose+(ipClose-ipLow+ipAtr)*multi)
if strategy.position_size < 0
    strategy.exit("S", stop=ipHigh+ipAtr, limit=ipClose-(ipHigh-ipClose+ipAtr)*multi)



// plot(ipAtr)
// plot(ipClose)
// plot(ipHigh)
// plot(ipLow)
    

```
