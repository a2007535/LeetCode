```python
class Solution:
    def __init__(self,nums) -> None:
        self.nums=nums
    def reverse(self,firstIndex, lastIndex):
        first=firstIndex
        last=lastIndex
        temp = 0
        while last > first:
            temp=self.nums[last]
            self.nums[last]=self.nums[first]
            self.nums[first] = temp
            last -=1
            first +=1

    def moveZeroes(self) -> None:
        firstIndex = 0
        lastIndex = len(self.nums)-1
        while lastIndex > firstIndex :
            if self.nums[firstIndex] == 0:
                self.reverse(firstIndex,lastIndex)
                lastIndex-=1
                self.reverse(firstIndex,lastIndex)
            firstIndex +=1
        """
        Do not return anything, modify nums in-place instead.
        """

        
```

用 two pointer 找到 第一個0後 將 他跟 最後一個number做reverse
step one : [0,1,0,3,12] 第一個跟最後一個reverse => [12,3,0,1,0] 然後 最後一位 往前一位 變成倒數第二位 將 找到0的那個位置跟 倒數第二位做 reverse [1,0,3,12,0] 這樣就可以不改變順序 變換0的位置了
step two : [1,0,3,12,0] => [1,12,3,0,0] => [1,3,12,0,0]
